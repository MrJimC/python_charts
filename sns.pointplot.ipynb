```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("data_from_trip_practice_only.csv")
df.head()
```

## Normal use

**默認 `estimator="mean"`**

```python
sns.set_style("darkgrid") # 灰色網格背景
sns.pointplot(x="countrynameen", y="this_wk_uv", data=df)
```

圖中的點為這組數據的平均值點，豎線則為誤差棒，默認兩個均值點會相連接，若不想顯示，可以通過`join`參數實現：

## Other parameters

### `join`

```python
sns.pointplot(x="countrynameen", y="this_wk_uv", data=df, join=False)
```

### `hue`

```python
sns.pointplot(x="countrynameen", y="this_wk_uv", data=df, hue="lose_rate_tag")
```

我們可以看到兩個類別的誤差棒重疊在了一起，使數據觀測不清晰。<br>
可以使用pointplot的`dodge`參數將重疊的部分錯開：

### `dodge`

```python
sns.pointplot(x="countrynameen", y="this_wk_uv", data=df, hue="lose_rate_tag", dodge=True)
```

### `markets` , `linestyles`

```python
sns.pointplot(x="countrynameen", y="this_wk_uv", data=df, hue="lose_rate_tag", dodge=True
              , markers=["*","x","*"], linestyles=["-.","--","--"])
```
