```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("data_from_trip_practice_only.csv")
df.head()
```

```python
competitive = df.competitive_tag.value_counts()
competitive
```

```python
fig = plt.figure(figsize = (10,5))
plt.bar(competitive.index, competitive, width=0.4, color='g')

# plt.bar(competitive.index, competitive.values, width=0.4, color='g') ## this could work as well

plt.xlabel("competitive tags")
plt.ylabel("No of hotels")
plt.title('Number of hotels in each tag',color = 'black',fontsize = 20)
plt.show()
```
